```python
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression

import pandas as pd
import numpy as np
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
fluai = pd.read_excel('agz.xlsx')
fluai = fluai.drop(['Kişi'],axis=1)
```

```python
fluai
```

```python
fluai.corr(method ='pearson') 
```

```python
X = fluai.drop(['Health_point'],axis=1)
```

```python
y = fluai['Health_point']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

### Linear Regression

```python
model = LinearRegression()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
list(zip(X_train.columns, model.coef_)) 
```

A Linear Regression achieves an R^2 coefficient of around 0.7.
